```python
# ==========================================
# PIPELINE DATA CLEANING - HOUSING DATASET
# ==========================================

# Import Library
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

# ==========================================
# STEP 0 - Load Dataset & Eksplorasi Awal
# ==========================================

# Upload file housing_dirty.csv terlebih dahulu ke Colab
df = pd.read_csv('housing_dirty.csv')

print("=== Shape Awal Dataset ===")
print(df.shape)

print("\n=== Informasi Dataset ===")
df.info()

print("\n=== Missing Values Awal ===")
print(df.isnull().sum())

print("\n=== Statistik Deskriptif ===")
print(df.describe())


# ==========================================
# STEP 1 - Menghapus Data Duplikat
# ==========================================

df.drop_duplicates(inplace=True)

print("\n=== Setelah Hapus Duplikat ===")
print(df.shape)


# ==========================================
# STEP 2 - Normalisasi String
# ==========================================

# Kolom kota
df['kota'] = df['kota'].str.strip().str.title()

# Kolom kondisi
df['kondisi'] = df['kondisi'].str.strip().str.lower()

print("\n=== Data Setelah Normalisasi String ===")
print(df[['kota', 'kondisi']].head())


# ==========================================
# STEP 3 - Imputasi Missing Values
# ==========================================

# Median untuk numerik
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())

# Modus untuk kategorik
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])

print("\n=== Missing Values Setelah Imputasi ===")
print(df.isnull().sum())


# ==========================================
# STEP 4 - Menangani Outlier (IQR Fence)
# ==========================================

for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower_fence, upper_fence)

print("\n=== Outlier Berhasil Ditangani ===")


# ==========================================
# STEP 5 - Validasi Dataset
# ==========================================

print("\n=== Validasi Dataset ===")

print("Jumlah Missing Values:")
print(df.isnull().sum())

print("\nJumlah Data Duplikat:")
print(df.duplicated().sum())


# ==========================================
# STEP 6 - Export Dataset Bersih
# ==========================================

df.to_csv('housing_clean.csv', index=False)

print("\nDataset bersih berhasil disimpan!")
print("Nama file: housing_clean.csv")


# ==========================================
# STEP 7 - API JSONPlaceholder
# ==========================================

url = "https://jsonplaceholder.typicode.com/users"

api_df = pd.read_json(url)

print("\n=== Data dari API JSONPlaceholder ===")
print(api_df.head())
```
