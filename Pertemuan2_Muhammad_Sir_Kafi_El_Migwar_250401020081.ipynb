```python
# =========================
# Import Library
# =========================
import pandas as pd
import numpy as np


# =========================
# Load Dataset
# =========================
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset berhasil dimuat")


# =========================
# Menampilkan Data Awal
# =========================
print("\n=== 5 Data Teratas ===")
print(df.head())


# =========================
# Informasi Dataset
# =========================
print("\n=== Informasi Dataset ===")
df.info()


# =========================
# Ukuran Dataset
# =========================
print("\n=== Jumlah Baris dan Kolom ===")
print(df.shape)


# =========================
# Nama Kolom
# =========================
print("\n=== Nama Kolom ===")
print(df.columns)


# =========================
# Missing Values
# =========================
print("\n=== Cek Missing Values ===")
print(df.isnull().sum())


# =========================
# Statistik Deskriptif
# =========================
print("\n=== Statistik Deskriptif ===")
print(df.describe())


# =========================
# Q1: Jumlah baris, kolom, dan nama kolom
# =========================
print("\n=== Q1 ===")
print("Jumlah baris dan kolom:", df.shape)
print("Nama kolom:", df.columns)


# =========================
# Q2: Missing values + persentase
# =========================
print("\n=== Q2 ===")
missing = df.isnull().sum()
persen = (missing / len(df)) * 100
print("Jumlah missing value:\n", missing)
print("\nPersentase missing value (%):\n", persen)


# =========================
# Q3: Persentase penumpang selamat
# =========================
print("\n=== Q3 ===")
survival_rate = df['Survived'].mean() * 100
print("Persentase penumpang selamat:", survival_rate, "%")


# =========================
# Q4: Wanita kelas 1
# =========================
print("\n=== Q4 ===")
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print("Jumlah wanita kelas 1:", len(wanita_kelas1))
print("Rata-rata usia:", wanita_kelas1['Age'].mean())


# =========================
# Q5: Survival per kelas
# =========================
print("\n=== Q5 ===")
survival_class = df.groupby('Pclass')['Survived'].mean()
print(survival_class)
print("Kelas dengan survival tertinggi:", survival_class.idxmax())


# =========================
# Q6: Analisis tambahan
# =========================
print("\n=== Q6 ===")
survival_gender = df.groupby('Sex')['Survived'].mean()
print("Survival berdasarkan gender:\n", survival_gender)
```
